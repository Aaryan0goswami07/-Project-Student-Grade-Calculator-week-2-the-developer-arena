```python
# grade_calculator.py
# Student Grade Calculator - Week 2 Project

def get_grade(marks):
    """Returns grade based on marks."""
    if 90 <= marks <= 100:
        return "A"
    elif 80 <= marks <= 89:
        return "B"
    elif 70 <= marks <= 79:
        return "C"
    elif 60 <= marks <= 69:
        return "D"
    else:
        return "F"


def get_message(grade):
    """Returns encouraging message based on grade."""
    if grade == "A":
        return "Excellent! You're doing amazing! ğŸŒŸ"
    elif grade == "B":
        return "Very Good! Keep it up! ğŸ‘"
    elif grade == "C":
        return "Good effort! You can improve even more! ğŸ’ª"
    elif grade == "D":
        return "Don't give up! Keep practicing and you'll grow! ğŸ“ˆ"
    else:
        return "It's okay! Failures help you learn â€” bounce back stronger! ğŸ”¥"


# --- Main Program ---
print("ğŸ¯ Student Grade Calculator ğŸ¯")
print("-" * 40)

student_name = input("Enter student name: ").strip()

# Validation loop
while True:
    marks_input = input("Enter marks (0-100): ").strip()

    # Check if input is numeric
    if marks_input.isdigit():
        marks = int(marks_input)

        # Check valid range
        if 0 <= marks <= 100:
            break
        else:
            print("âŒ Invalid marks. Please enter marks between 0 and 100.")
    else:
        print("âŒ Invalid input. Please enter numbers only.")

# Grading
grade = get_grade(marks)
message = get_message(grade)

# Output result
print("\n" + "=" * 50)
print(f"ğŸ“Š RESULT FOR {student_name.upper()}:")
print(f"Marks: {marks}/100")
print(f"Grade: {grade}")
print(f"Message: {message}")
print("=" * 50)
```
